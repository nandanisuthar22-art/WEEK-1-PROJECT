```python
# SPORTS DATA ANALYSIS PROJECT

 # 1.Import necessary libraries

import pandas as pd
from google.colab import files
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')
```

```python
# 2. DATA LOADING
df = pd.read_csv('/content/drive/MyDrive/sports_dataset.csv')
print("="*60)
print("DATA LOADING - Initial Dataset")
print("="*60)
print(f"Dataset Shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
print("\nFirst 5 rows:")
print(df.head())
```

```python
# 3. DATA CLEANING
print("\n" + "="*60)
print("DATA CLEANING")
print("="*60)

# Create a copy for cleaning
df_clean = df.copy()

# 2.1 Handle missing values
print("\n2.1 Missing Values Before Cleaning:")
print(df_clean.isnull().sum())

# Fill missing Player Names with 'Unknown'
df_clean['Player Name'] = df_clean['Player Name'].fillna('Unknown')

# Fill missing team with 'Unknown Team'
df_clean['team'] = df_clean['team'].fillna('Unknown Team')

# Fill missing match_date with a placeholder
df_clean['match_date'] = df_clean['match_date'].fillna('01-01-2026')

```

```python
# 2.2 Clean and convert numeric columns
# Clean sprints column (replace 'twenty', 'NaN', etc. with numeric)
df_clean['sprints'] = df_clean['sprints'].replace(['twenty', 'NaN', 'nan', ''], np.nan)
df_clean['sprints'] = pd.to_numeric(df_clean['sprints'], errors='coerce')

# Clean top_speed column
df_clean['top_speed(km/h)'] = df_clean['top_speed(km/h)'].replace(['invalid', 'NaN', 'nan', ''], np.nan)
df_clean['top_speed(km/h)'] = pd.to_numeric(df_clean['top_speed(km/h)'], errors='coerce')

# Clean distance_km column
df_clean['distance_km'] = pd.to_numeric(df_clean['distance_km'], errors='coerce')

# Clean goals column
df_clean['goals'] = pd.to_numeric(df_clean['goals'], errors='coerce').fillna(0)

# Clean assists column
df_clean['assists'] = df_clean['assists'].replace(['', 'NaN', 'nan'], np.nan)
df_clean['assists'] = pd.to_numeric(df_clean['assists'], errors='coerce').fillna(0)

# Clean minutes_played column (replace 'eighty', 'eighty', etc.)
def convert_minutes(val):
    if pd.isna(val):
        return np.nan
    if isinstance(val, str):
        val = val.lower()
        if 'eighty' in val or val == 'eighty':
            return 80
        if val == 'ninety':
            return 90
    return pd.to_numeric(val, errors='coerce')

df_clean['minutes_played'] = df_clean['minutes_played'].apply(convert_minutes)
df_clean['minutes_played'] = df_clean['minutes_played'].fillna(df_clean['minutes_played'].median())

# Clean result column (standardize)
def standardize_result(val):
    if pd.isna(val):
        return 'Draw'
    val = str(val).lower()
    if val in ['win', 'w']:
        return 'Win'
    elif val in ['loss', 'l', 'lost']:
        return 'Loss'
    else:
        return 'Draw'

df_clean['result'] = df_clean['result'].apply(standardize_result)
```

```python
# 2.3 Clean date column
def clean_date(date_str):
    try:
        if pd.isna(date_str):
            return pd.NaT
        date_str = str(date_str)
        # Handle different date formats
        if '/' in date_str:
            return pd.to_datetime(date_str, format='%Y/%m/%d', errors='coerce')
        elif '-' in date_str:
            return pd.to_datetime(date_str, format='%d-%m-%Y', errors='coerce')
        else:
            return pd.to_datetime(date_str, errors='coerce')
    except:
        return pd.NaT

df_clean['match_date'] = df_clean['match_date'].apply(clean_date)
```

```python
# 2.4 Remove duplicates
df_clean = df_clean.drop_duplicates(subset=['player_id', 'match_date'], keep='first')
```

```python
# 2.5 Handle outliers in numeric columns
for col in ['sprints', 'top_speed(km/h)', 'distance_km', 'goals', 'assists', 'minutes_played']:
    if col in df_clean.columns:
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df_clean[col] = df_clean[col].clip(lower=lower_bound, upper=upper_bound)
```

```python
print("\n2.6 Missing Values After Cleaning:")
print(df_clean.isnull().sum())
```

```python
print("\n2.7 Dataset Info After Cleaning:")
print(df_clean.info())
```

```python
# 4. DATA VISUALIZATIONS ( with Hover Effects)

print("\n" + "="*60)
print("GENERATING VISUALIZATIONS")
print("="*60)
```

```python

# Visualization 1: Distribution of Goals with Hover (Plotly)
print("\n📊 Visualization 1: Distribution of Goals - Interactive")
fig1 = px.histogram(df_clean, x='goals',
                     title='Distribution of Goals Scored by Players',
                     labels={'goals': 'Number of Goals', 'count': 'Number of Players'},
                     color_discrete_sequence=['#2E86AB'],
                     hover_data=['player_id', 'Player Name', 'team'])
fig1.update_layout(title_font_size=20, title_x=0.5, bargap=0.2)
fig1.show()
```

```python
# Visualization 2: Top Speed Distribution by Team (Box Plot with Hover)
print("\n📊 Visualization 2: Top Speed Distribution by Team")
fig2 = px.box(df_clean.dropna(subset=['top_speed(km/h)']),
               x='team', y='top_speed(km/h)',
               color='team',
               title='Top Speed Distribution Across Teams',
               labels={'top_speed(km/h)': 'Top Speed (km/h)', 'team': 'Team'},
               hover_data=['player_id', 'Player Name', 'result'])
fig2.update_layout(title_font_size=20, title_x=0.5, showlegend=False)
fig2.show()
```

```python
# Visualization 3: Scatter Plot - Distance vs Sprints (Colored by Result)
print("\n📊 Visualization 3: Distance vs Sprints Performance Analysis")
filtered_df = df_clean.dropna(subset=['distance_km', 'sprints'])
fig3 = px.scatter(filtered_df, x='distance_km', y='sprints',
                   color='result', size='goals',
                   hover_data=['player_id', 'Player Name', 'team', 'top_speed(km/h)'],
                   title='Distance Covered vs Sprints (Size = Goals Scored)',
                   labels={'distance_km': 'Distance Covered (km)', 'sprints': 'Number of Sprints'})
fig3.update_layout(title_font_size=20, title_x=0.5)
fig3.show()
```

```python
# Visualization 4: Goals and Assists by Team (Bar Chart)
print("\n📊 Visualization 4: Team Performance - Goals and Assists")
team_stats = df_clean.groupby('team')[['goals', 'assists']].sum().reset_index()
team_stats_melted = team_stats.melt(id_vars='team', var_name='Metric', value_name='Count')
fig4 = px.bar(team_stats_melted, x='team', y='Count', color='Metric',
               barmode='group', title='Total Goals and Assists by Team',
               labels={'team': 'Team', 'Count': 'Total Count', 'value': 'Count'},
               hover_data={'Count': ':,.0f'})
fig4.update_layout(title_font_size=20, title_x=0.5)
fig4.show()
```

```python
# Visualization 5: Top 10 Players by Goals (Horizontal Bar Chart)
print("\n📊 Visualization 5: Top 10 Goal Scorers")
top_players = df_clean.groupby(['player_id', 'Player Name'])[['goals', 'assists']].sum().reset_index()
top_players = top_players.nlargest(10, 'goals')
fig5 = px.bar(top_players, x='goals', y='Player Name',
               orientation='h', color='goals',
               text='goals', title='Top 10 Goal Scorers',
               labels={'goals': 'Total Goals', 'Player Name': 'Player'},
               color_continuous_scale='Viridis',
               hover_data={'assists': ':,.1f'})
fig5.update_traces(textposition='outside')
fig5.update_layout(title_font_size=20, title_x=0.5, height=500)
fig5.show()
```

```python
# Visualization 6: Performance by Day of Week (Line Chart)
print("\n📊 Visualization 6: Average Goals by Day of Week")
df_clean['day_of_week'] = df_clean['match_date'].dt.day_name()
daily_performance = df_clean.groupby('day_of_week').agg({
    'goals': 'mean',
    'sprints': 'mean',
    'distance_km': 'mean'
}).reset_index()
# Order days correctly
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
daily_performance['day_of_week'] = pd.Categorical(daily_performance['day_of_week'], categories=day_order, ordered=True)
daily_performance = daily_performance.sort_values('day_of_week')
fig6 = px.line(daily_performance, x='day_of_week', y='goals',
                title='Average Goals Scored by Day of Week',
                labels={'goals': 'Average Goals', 'day_of_week': 'Day of Week'},
                markers=True, line_shape='linear',
                hover_data={'sprints': ':,.1f', 'distance_km': ':,.2f'})
fig6.update_traces(line=dict(width=3, color='#E63946'), marker=dict(size=10))
fig6.update_layout(title_font_size=20, title_x=0.5)
fig6.show()
```

```python
# Visualization 7: Correlation Heatmap (Matplotlib + Seaborn)
print("\n📊 Visualization 7: Correlation Heatmap of Performance Metrics")
numeric_cols = ['sprints', 'top_speed(km/h)', 'distance_km', 'goals', 'assists', 'minutes_played']
corr_matrix = df_clean[numeric_cols].corr()
plt.figure(figsize=(10, 8))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, mask=mask, annot=True, cmap='coolwarm', center=0,
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Correlation Matrix of Performance Metrics', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Visualization 8: Sunburst Chart - Team Hierarchy (Plotly)
print("\n📊 Visualization 8: Team Performance Sunburst Chart")
team_result_stats = df_clean.groupby(['team', 'result']).size().reset_index(name='count')
fig8 = px.sunburst(team_result_stats, path=['team', 'result'], values='count',
                    title='Match Results Distribution by Team',
                    color='count', color_continuous_scale='RdBu',
                    hover_data={'count': ':,.0f'})
fig8.update_layout(title_font_size=20, title_x=0.5)
fig8.show()
```

```python
# Visualization 9: Treemap - Player Performance
print("\n📊 Visualization 9: Player Performance Treemap")
player_perf = df_clean.groupby('Player Name').agg({
    'goals': 'sum',
    'assists': 'sum',
    'player_id': 'count'
}).reset_index()
player_perf['total_contributions'] = player_perf['goals'] + player_perf['assists']
player_perf = player_perf.nlargest(20, 'total_contributions')
fig9 = px.treemap(player_perf, path=['Player Name'], values='total_contributions',
                   color='goals', color_continuous_scale='Greens',
                   title='Top 20 Players by Total Contributions (Goals + Assists)',
                   hover_data={'assists': ':,.0f', 'goals': ':,.0f'})
fig9.update_layout(title_font_size=20, title_x=0.5)
fig9.show()
```

```python
# Visualization 10: Polar Chart - Player Attributes
print("\n📊 Visualization 10: Player Performance Radar Chart")
# Get top 5 players by goals
top5_players = df_clean.groupby('Player Name')[['goals', 'assists', 'sprints', 'top_speed(km/h)']].mean().reset_index()
top5_players = top5_players.nlargest(5, 'goals')
categories = ['goals', 'assists', 'sprints', 'top_speed(km/h)']
fig10 = go.Figure()
for _, player in top5_players.iterrows():
    fig10.add_trace(go.Scatterpolar(
        r=[player[cat] if pd.notna(player[cat]) else 0 for cat in categories],
        theta=categories,
        fill='toself',
        name=player['Player Name']
    ))
fig10.update_layout(
    polar=dict(radialaxis=dict(visible=True, range=[0, max(top5_players['goals'].max(), 5)])),
    title='Top 5 Goal Scorers - Performance Radar Chart',
    title_font_size=20,
    title_x=0.5,
    showlegend=True
)
fig10.show()
```

```python
# Visualization 11: Area Chart - Cumulative Performance Over Time
print("\n📊 Visualization 12: Cumulative Goals Over Time")
df_time = df_clean.sort_values('match_date')
df_time['cumulative_goals'] = df_time['goals'].cumsum()
df_time['cumulative_assists'] = df_time['assists'].cumsum()
fig12 = px.area(df_time, x='match_date', y=['cumulative_goals', 'cumulative_assists'],
                 title='Cumulative Goals and Assists Over Time',
                 labels={'value': 'Cumulative Count', 'match_date': 'Match Date',
                         'variable': 'Metric'},
                 color_discrete_map={'cumulative_goals': '#2E86AB', 'cumulative_assists': '#E63946'},
                 hover_data={'player_id': True, 'team': True})
fig12.update_layout(title_font_size=20, title_x=0.5)
fig12.show()
```

```python
# Team Rankings
team_rankings = df_clean.groupby('team').agg({
    'goals': 'sum',
    'assists': 'sum',
    'player_id': 'count'
}).rename(columns={'player_id': 'matches_played'})
team_rankings['total_points'] = team_rankings['goals'] + team_rankings['assists'] * 0.5
team_rankings = team_rankings.sort_values('total_points', ascending=False)

print("\n🏆 Team Performance Rankings:")
print(team_rankings.to_string())
```
